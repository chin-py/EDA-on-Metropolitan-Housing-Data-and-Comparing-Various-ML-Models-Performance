```python
# df["Area"].eq(9).sum() # is 0
df=data.replace(9,np.nan)
```

```python
# Upon the analysis of data we find that presence of 9(or NaN after replacing) is in a manner that at a time 30+ features are NaN.
# Hence in such case we'll drop the rows instead of doing fillna() or using SimpleImputer 
df.dropna(inplace=True)
df.info()
```

```python
df.isna().sum()
df.shape
```

```python
# Modifying price to price in lakhs(INR)
df['Price'] = df['Price']/100000
```

```python
numeric_cols=df.select_dtypes(exclude="object")
# or
# data.select_dtypes(include=['int','float']) # all except location attribute are of int type 
# numeric_cols.corr()
# numeric_cols.corr()
plt.figure(figsize=(30,30))
sns.heatmap(numeric_cols.corr(), annot=True, square=True)
```

```python
# low_corr_pairs = numeric_cols.corr().abs() < 0.9
# low_corr_pairs
```

```python
# Dropping the features based on the domain knowledge backed by correlation analysis
low_corr_cols=['Wifi','VaastuCompliant','Intercom','ShoppingMall','LandscapedGardens','StaffQuarter','Cafeteria','MultipurposeRoom','Hospital','ATM']

df.drop(low_corr_cols,axis=1,inplace=True)
df.info()
```

```python
from sklearn.decomposition import PCA
# Columns representing furnishing status
furnishing_cols = ['WashingMachine', 'Gasconnection', 'AC', 'Microwave','BED','TV', 'DiningTable', 'Sofa', 'Wardrobe', 'Refrigerator']
# Apply PCA
pca = PCA(n_components=1)
df['furnished_pca'] = pca.fit_transform(df[furnishing_cols])
df = df.drop(columns=furnishing_cols)
df.info()
```

```python
df['Area'].describe()
```

```python
# The Location column is of non numeric type and needs to be converted into numeric to build the model
# We will now take two separate approaches to prepare the data for model building based on model's suitability.
# For model like linear regression we'll Standardize required column and perform one-hot encoding on "Location" col.
# For  tree based models like Decision Trees, Random Forest, XGBoost we'll perform Label Encoding on "Location" col.

df['Location'] = df['Location'].str.strip().str.lower()

df['Location'].value_counts()
```

```python
# Approach 1, for models like Linear Regression, logistic regression
# One-hot encoding
# Standardizing the area
# There are 100+ unique Localities in this city, Hence grouping/renaming the ones having fewer records as "Other_location"
location_counts = df['Location'].value_counts()
rare_locs = location_counts[location_counts < 15].index # .index gets just the location names (index) of rare locations(filtered locations)
df['Location'] = df['Location'].apply(lambda x: 'other' if x in rare_locs else x) # Applies the lambda function to every value in the Location column
# df['Location'].value_counts()
df['Location'].nunique()
```

```python
# Performing One-Hot encoding for one set of data
df_ohe=pd.get_dummies(df, columns=['Location'], drop_first=True)
```

```python
df_ohe.shape # (1951, 73)
df_ohe.dtypes
```

```python
# Columns created by OHE are of type bool, need to convert them to type same as other columns
# we'll conver to float as other numeric catagorical are also float 0.0 and 1.0
bool_columns = df_ohe.select_dtypes(include=['bool']) # Selects only columns with boolean data type
bool_column_names = df_ohe.select_dtypes(include=['bool']).columns #Gets just the column names as an Index
type_mapping = {col: float for col in df_ohe.select_dtypes(include=['bool']).columns} # Creating dictionary mapping each boolean column name to float type
df_ohe = df_ohe.astype(type_mapping)
df_ohe.info()
```

```python
# Scaling the area(per sq. feet column)
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
df_ohe['Area_scaled'] = scale.fit_transform(df_ohe[['Area']])
df_ohe.drop(['Area'],axis=1,inplace=True)
df_ohe['Area_scaled'].describe()
```

```python
#Separating the independent and dependent variable
X1=df_ohe.drop('Price' ,axis=1)
Y1=df_ohe['Price']
```

```python
# Splitting the data into train and test set for model training and testing
from sklearn.model_selection import train_test_split
x_train1, x_test1, y_train1, y_test1 = train_test_split(X1, Y1, train_size  = 0.8, random_state = 123456)
```

```python
# 1) linear regression model
from sklearn.linear_model import LinearRegression
model_lr = LinearRegression()
model_lr.fit(x_train1, y_train1)

#Predicting
y_pred_lr=model_lr.predict(x_test1)
```

```python
# Evaluation metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score,root_mean_squared_error

mae = mean_absolute_error(y_test1, y_pred_lr)
mse = mean_squared_error(y_test1, y_pred_lr)
rmse = root_mean_squared_error(y_test1, y_pred_lr)
r2 = r2_score(y_test1, y_pred_lr)  # r2 is same as model score

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared (RÂ²):", r2)
```

```python
# 2) Support Vector Regression SVR
from sklearn.svm import SVR
model_svr = SVR(C=15)  # C parameter controls the strength of the regularization applied, higher C lower regularization
model_svr.fit(x_train1, y_train1)

#Predicting
y_pred_svr=model_svr.predict(x_test1)
```

```python
# Evaluation metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score,root_mean_squared_error

mae = mean_absolute_error(y_test1, y_pred_svr)
mse = mean_squared_error(y_test1, y_pred_svr)
rmse = root_mean_squared_error(y_test1, y_pred_svr)
r2 = r2_score(y_test1, y_pred_svr)  # r2 is same as model score

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared (RÂ²):", r2)
```

```python
# 3) Lasso Regression
from sklearn.linear_model import Lasso

model_lasso = Lasso(alpha=30) # adds the sum of the absolute values of the coefficients to the loss function, encouraging sparsity(feature selection) by driving some coefficients to exactly zero.
# Regularization strength (higher = more regularization)
model_lasso.fit(x_train1, y_train1)
y_pred_lasso=model_lasso.predict(x_test1)
```

```python
#  Evaluation metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score,root_mean_squared_error

mae = mean_absolute_error(y_test1, y_pred_lasso)
mse = mean_squared_error(y_test1, y_pred_lasso)
rmse = root_mean_squared_error(y_test1, y_pred_lasso)
r2 = r2_score(y_test1, y_pred_lasso)  # r2 is same as model score

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared (RÂ²):", r2)
```

```python
# 4) Ridge Regression
from sklearn.linear_model import Ridge

model_ridge = Ridge(alpha=15) # penalizes the square of the magnitude of the model's coefficients #(higher = more regularization)
model_ridge.fit(x_train1, y_train1)
y_pred_ridge=model_ridge.predict(x_test1)
```

```python
#  Evaluation metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score,root_mean_squared_error

mae = mean_absolute_error(y_test1, y_pred_ridge)
mse = mean_squared_error(y_test1, y_pred_ridge)
rmse = root_mean_squared_error(y_test1, y_pred_ridge)
r2 = r2_score(y_test1, y_pred_ridge)  # r2 is same as model score

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared (RÂ²):", r2)
```

```python
# 5) Elastic Net (Combines L1 and L2 penalties of Lasso and Ridge respt.)
from sklearn.linear_model import ElasticNet

model_elastic = ElasticNet(alpha=1, l1_ratio=0.6)  # l1_ratio controls L1 vs L2 mix
model_elastic.fit(x_train1, y_train1)
y_pred_elastic=model_elastic.predict(x_test1)
```

```python
#  Evaluation metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score,root_mean_squared_error

mae = mean_absolute_error(y_test1, y_pred_elastic)
mse = mean_squared_error(y_test1, y_pred_elastic)
rmse = root_mean_squared_error(y_test1, y_pred_elastic)
r2 = r2_score(y_test1, y_pred_elastic)  # r2 is same as model score

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared (RÂ²):", r2)
```

```python
# Approach 2 for Tree based Models
# Label encoding
# No scaling required:- These models split on thresholds, not distances or magnitudes, so they are scale-invariant.
# Handle Nonlinearity- Tree models learn nonlinear relationships very well.
#  Handle Categorical Features (like Location) Tree-based models donâ€™t assume any ordering or scaling of categorical variables. Hence label encoding is suitable choice
# Robust to Outliers Extreme price or area values donâ€™t affect tree splits much â€” unlike linear models.



from sklearn.preprocessing import LabelEncoder
df_le=df.copy()
le = LabelEncoder()
df_le['location_encoded'] = le.fit_transform(df_le['Location'])
# df_le['location_encoded'].describe() #max  44.000000 , min  0.000000
df_le.drop("Location",axis=1,inplace=True)
df_le.info()
```

```python
#Separating the independent and dependent variable
X2=df_le.drop('Price' ,axis=1)
Y2=df_le['Price']
```

```python
# Splitting the data into train and test set for model training and testing
from sklearn.model_selection import train_test_split
x_train2, x_test2, y_train2, y_test2 = train_test_split(X2, Y2, train_size  = 0.8, random_state = 123456)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
# Create all models
models = {
    'RandomForest': RandomForestRegressor(n_estimators=100, random_state=123,max_depth=80),
    'DecisionTree': DecisionTreeRegressor(random_state=123,min_samples_leaf=3,max_depth=100),
    'GradientBoosting': GradientBoostingRegressor(n_estimators=123, random_state=100),
    'XGBoost': XGBRegressor(n_estimators=100, random_state=123)
}

predictions = {}

# Train all models
for name, model in models.items():
    model.fit(x_train2, y_train2)
    predictions[name] = model.predict(x_test2)  # Store predictions by model name
    # print(f"{name} model trained successfully!")
```

```python
#  Evaluation metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score,root_mean_squared_error

for name, model in models.items():
    mae = mean_absolute_error(y_test1, predictions[name])
    mse = mean_squared_error(y_test1, predictions[name])
    rmse = root_mean_squared_error(y_test1, predictions[name])
    r2 = r2_score(y_test1, predictions[name])  # r2 is same as model score
    
    print(f"Model: {name:^30}")
    print("Mean Absolute Error (MAE):", mae)
    print("Mean Squared Error (MSE):", mse)
    print("Root Mean Squared Error (RMSE):", rmse)
    print("R-squared (RÂ²):", r2)
    print("-"*50)
    print()
```
